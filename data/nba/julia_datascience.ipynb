```julia
df1 = CSV.read(HTTP.get("https://raw.githubusercontent.com/prehensilecode/julia-beginners/master/data/s_p500companies/constituents_csv.csv").body, DataFrame)
df1 = Impute.interp(df1)
```

```julia
# sorting
sort!(df, rev=false)
```

```julia
sort!(df, rev=true)
```

```julia
# split data: train, test, validation
using MLDataUtils
df2, df3 = splitobs(shuffleobs(df), at=(0.5, 0.4))
```

```julia
# advanced joins
# inner join = intersection
# full outer inclusive join = union
# left inclusive join = (A \ B) union (A intersect B)
# right inclusive join = (B \ A) union (B intersect A)
df4 = innerjoin(df1, df, on=:Name, makeunique=true)
```

```julia
df5 = outerjoin(df1, df, on=:Name, makeunique=true)
```

```julia
df7 = leftjoin(df1, df, on=:Name, makeunique=true)
```

```julia
df8 = rightjoin(df1, df, on=:Name, makeunique=true)
```
